```python
# importing required libraries
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

```python
my_url = 'https://www.imdb.com/list/ls086535241/?sort=num_votes,desc&st_dt=&mode=detail&page=1'
uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()
```

```python
page_html
```

```python
# parser for html
page_soup = soup(page_html, 'html.parser')
S = soup(uClient)
```

```python
page_soup
```

```python
containers = page_soup.findAll('div', {'class': 'lister-item mode-detail'})
print(len(containers))
```

```python
print(soup.prettify(containers[0]))
```

```python
filename = 'imdb_m.csv'
f = open(filename, 'w')

headers = 'Name, Year, Runtime\n'
f.write(headers)

for container in containers:
    name = container.img['alt']
    mov_year = container.findAll('span', {'class': 'lister-item-year'})
    year = mov_year[0].text
    mov_runtime = container.findAll('span', {'class': 'runtime'})
    runtime = mov_runtime[0].text
    
    print(name + ',' + year + ',' + runtime + '\n')
    f.write(name + ',' + year + ',' + runtime + '\n')

f.close()
```
